```python
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Samet\Desktop\diabetes2.csv")  #verileri yüklüyoruz
```

```python
data.shape  #verisetimizin boyutunu bastırıyoruz
```

```python
new_data = data.copy() #orjinal verisetimizin bozulmaması için verisetimizi bir başka değişkene kopyalıyoruz ve işlemlerimizi onun üerinden yapıyoruz.
```

```python
x=new_data.drop("Outcome",axis = 1)  #bağımsız değişkenleri x adlı değişkene atamak için 1 tane olan bağımlı değişkeni çıkartıyoruz
y= new_data["Outcome"]#bağımlı değişkenimizi y isimli değişkene atıyoruz.
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.20) #verilerimizi %20 oranında train ve test diye ayırıyoruz.
```

```python
import seaborn as sns
korelasyon = new_data.corr() #veriler arasındaki korelasyonu belirliyoruz.
sns.heatmap(korelasyon,annot=True,linewidth = 1)  #heatmap grafiğini bastırıyoruz.
```

```python
predict = lr.predict(x_test) #modelin test verilerine göre tahminlerini alıyoruz.
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(predict,y_test) #comfusion matrix ile modelin doğruluğunu test ediyoruz.
```
